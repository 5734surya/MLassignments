```python
#1. Implement a new class "Agent"  with following attributes - name, alive , action 
#2. define constructors of the class "Agent" 
#3. define necessary methods in the class "Agent"

class Agent:
    def __init__(self, name, alive, action):
        self.name = name
        self.alive = alive
        self.action = action

    def __str__(self):
        return "Name: {}\n Alive: {}\n Action: {}".format(self.name, self.alive, self.action)

    def do_action(self):
        if self.action == "shoot":
            print(self.name + " is shooting!")
        elif self.action == "run":
            print(self.name + " is running!")
        else:
            print(self.name + " is doing something else!")


a = Agent("John", True, "shoot")
print(a)
a.do_action()

#Implement a superclass "Vacuum" of the class "Agent". Define constructor(s), necessary attributes, and methods. 


class Vacuum(Agent):
    def __init__(self, name, alive, action, power):
        super().__init__(name, alive, action)
        self.power = power

    def __str__(self):
        return super().__str__() + "\n Power: {}".format(self.power)

    def do_action(self):
        if self.action == "shoot":
            print(self.name + " is shooting!")
        elif self.action == "run":
            print(self.name + " is running!")
        elif self.action == "vacuum":
            print(self.name + " is vacuuming!")
        else:
            print(self.name + " is doing something else!")


v = Vacuum("Bobby", True, "vacuum", 100)
print(v)
v.do_action()

# Implement a new class "Environment" that model the vacuum world with two rooms. Each room has two possible statuses: clean and dirty.

class Environment:
    def __init__(self, room1_status, room2_status):
        self.room1_status = room1_status
        self.room2_status = room2_status

    def __str__(self):
        return "Room 1 Status: {}\n Room 2 Status: {}".format(self.room1_status, self.room2_status)

    def is_dirty(self):
        if self.room1_status == "dirty" or self.room2_status == "dirty":
            return True
        else:
            return False

    def clean_room(self, room_number):
        if room_number == 1:
            self.room1_status = "clean"
        elif room_number == 2:
            self.room2_status = "clean"
        else:
            print("Invalid room number!")


e = Environment("clean", "dirty")
print(e)
print(e.is_dirty())
e.clean_room(2)
print(e)

#Write a test program that tests these classes, methods, and print the status of agent and environment.

class Agent:
    def __init__(self, name, alive, action):
        self.name = name
        self.alive = alive
        self.action = action

    def __str__(self):
        return "Name: {}\n Alive: {}\n Action: {}".format(self.name, self.alive, self.action)

    def do_action(self):
        if self.action == "shoot":
            print(self.name + " is shooting!")
        elif self.action == "run":
            print(self.name + " is running!")
        else:
            print(self.name + " is doing something else!")


class Vacuum(Agent):
    def __init__(self, name, alive, action, power):
        super().__init__(name, alive, action)
        self.power = power

    def __str__(self):
        return super().__str__() + "\n Power: {}".format(self.power)

    def do_action(self):
        if self.action == "shoot":
            print(self.name + " is shooting!")
        elif self.action == "run":
            print(self.name + " is running!")
        elif self.action == "vacuum":
            print(self.name + " is vacuuming!")
        else:
            print(self.name + " is doing something else!")


class Environment:
    def __init__(self, room1_status, room2_status):
        self.room1_status = room1_status
        self.room2_status = room2_status

    def __str__(self):
        return "Room 1 Status: {}\n Room 2 Status: {}".format(self.room1_status, self.room2_status)

    def is_dirty(self):
        if self.room1_status == "dirty" or self.room2_status == "dirty":
            return True
        else:
            return False

    def clean_room(self, room_number):
        if room_number == 1:
            self.room1_status = "clean"
        elif room_number == 2:
            self.room2_status = "clean"
        else:
            print("Invalid room number!")


a = Agent("John", True, "shoot")
print(a)
a.do_action()

v = Vacuum("Bobby", True, "vacuum", 100)
print(v)
v.do_action()

e = Environment("clean", "dirty")
print(e)
print(e.is_dirty())
e.clean_room(2)
print(e)
```
